```python
#Functions
def square(x):
    return x*x

def cube(x):
    return x**3

myList = [1, 2, 3, 4, 5]
for i in range(len(myList)):
    print(myList[i], end=" ")
    myList[i] = cube(myList[i])
    print(myList[i])
    
    
#can also do:
def do_a_thing(fn, the_list):
    result = []
    for value in the_list:
        result.append(fn(value))
    return result


print(do_a_thing(square, myList))


addOne = (lambda x : x-2)
print(do_a_thing(addOne, myList))
```
